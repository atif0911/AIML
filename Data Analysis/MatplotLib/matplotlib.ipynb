# Introduction to MatplotLib

### Importing pandas and PyPlot

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Example of Plotting using list

```python
x=[1,2,3,4,5]
plt.plot(x) #plots the relation x=x
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y) #plots  the relation x=y
```

### Naming the plot and axis

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y) #plots  the relation x=y
plt.title("Squares") #to give a title of the plot
plt.xlabel("This is x") #to label x axis
plt.ylabel("This is y") #to label y axis
plt.show() #shows the plotted relation
```

### More relations of x

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
z=[0,2,6,12,20]
plt.plot(x,y) #plots  the relation x=y
plt.plot(x,z) #plots  the relation x=z
plt.title("Squares") #to give a title of the plot
plt.xlabel("This is x axis") #to label x axis
plt.ylabel("This is y axis") #to label y axis
plt.show() #shows the plotted relation
```

### To label a relation

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
z=[0,2,6,12,20]
plt.plot(x,y) #plots  the relation x=y
plt.plot(x,z) #plots  the relation x=z
plt.title("Squares") #to give a title of the plot
plt.xlabel("This is x axis") #to label x axis
plt.ylabel("This is y axis") #to label y axis
plt.legend(['This is xRy','this is xRz']) #this takes a list as arg 
plt.show() #shows the plotted relation
```

### Compare population growth in india and china (Visualize)

```python
df.sample(10)
```

```python
ind=df.loc[df['country']=='India']
```

```python
ind
```

```python
china=df.loc[df['country']=='China']
```

```python
china
```

```python
plt.plot(ind['year'],ind['population']/10**6)
plt.plot(china['year'],china['population']/10**6)
plt.xlabel('Year')
plt.ylabel('Population (in Million)')
plt.legend(['India','China'])
plt.show()
```
