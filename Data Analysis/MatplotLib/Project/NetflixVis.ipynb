Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Reading CSV

```python
df=pd.read_csv('netflix_titles.csv')
```

```python
df.info()
```

```python
df.head(10)
```

## Clean data

```python
df=df.dropna(subset=['type','country','release_year','rating','duration']) #removing any NaN value
```

```python
df.head()
```

Bar Chart for Movies vs TV-Shows

```python
type_count=df['type'].value_counts()

plt.Figure(figsize=(6,4))
plt.bar(type_count.index,type_count.values,color=['skyblue','orange'])
plt.title('Number of movies vs shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()

#plt.savefig('movies_vs_shows.png')

plt.show()
```

Rating distributed

```python
rating_count=df['rating'].value_counts()

plt.figure(figsize=(8,6))

plt.pie(rating_count, labels=rating_count.index,autopct='%1.1f%%',startangle=90)
plt.title('Percentage of Rating')
plt.tight_layout()
plt.savefig('content_rating_pie.png')
plt.show()
```

```python
plt.bar(rating_count.index,rating_count.values,color=['skyblue','orange'])
plt.title('Type vs Rating')
plt.xlabel('Type')
plt.ylabel('Rating')
plt.tight_layout()

plt.savefig('content_rating_bar.png')

plt.show()
```

Duration of movies 

```python
movies_df=df.loc[df['type']=='Movie']
```

```python
movies_df['duration_int']=movies_df['duration'].str.replace(' min','').astype(int)

plt.Figure(figsize=(8,6))
plt.hist(movies_df['duration_int'],bins=30,color='purple',edgecolor='black')
plt.title('Distribution of Movie Duration')
plt.xlabel('Duration in mins')
plt.ylabel('Number of movies')
plt.tight_layout()

plt.savefig('movie_duration_hist.png')

plt.show()
```

Release Year vs Number of Shows

```python
release_count=df['release_year'].value_counts().sort_index()
plt.Figure(figsize=(10,6))
plt.scatter(release_count.index,release_count.values,color='red')
plt.title('Release Year VS Number of Shows')
plt.xlabel('Release Year')
plt.ylabel('Number of showss')
plt.tight_layout()

plt.savefig('release_year_scatter.png')

plt.show()
```

```python
content_by_year=df.groupby(['release_year','type']).size().unstack().fillna(0)
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,5))

#first subplot movies
ax[0].plot(content_by_year.index,content_by_year['Movie'],color='Skyblue')
ax[0].set_title('Movies Released Per Year')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Number of Movies')

#second plot shows
ax[1].plot(content_by_year.index,content_by_year['TV Show'],color='Orange')
ax[1].set_title('TV Shows Released Per Year')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Number of TV Shows')

fig.suptitle('Comparison of movies and tv shows over year')

plt.tight_layout()

plt.savefig('movie_show_comp.png')

plt.show()
```
