# Seaborn
Seaborn is a data visualisation library in python. It helps to create interesting and Interactive plots to draw meaningful information from data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Bar Plot
sns.barplot(x,y)
plt.show()

```python
df=pd.read_csv('netflix_titles.csv')
```

```python
df.head(5)
```

```python
df=df.dropna(subset=['release_year','rating','type','country','duration','director'])
```

No of movies per year

```python

res=sns.barplot(x=df['type'].value_counts().index,y=df['type'].value_counts().values,palette='pastel') #seaborn gives pallete or u can use hard color with color=''
plt.show()
```

COunt plot

```python
sns.countplot(x='type',data=df,palette='pastel')
```

```python
sns.countplot(x='director',data=df) #for horizontal use y axis
```

```python
sns.barplot(y='type',x='release_year',data=df)
```

```python
sns.displot(df['release_year'],bins=30,color='g')
```

```python
sns.displot(df['duration'],bins=30,color='g')
```

```python
iris = sns.load_dataset('iris') #inbuilt data set
```

```python
iris_n=iris.copy()
del iris_n['species']
```

```python
sns.heatmap(iris_n.corr(), cbar=True, linewidths=0.5)
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris)
```

LMPlot

```python
sns.lmplot(x='petal_length',y='petal_width',data=iris)
```

```python
sns.lmplot(x='petal_length',y='petal_width',data=iris,hue='species',markers=['o','^','*'])
```

boxPlot

```python
sns.boxplot(x='species',y='sepal_width',data=iris,hue='species')
```
